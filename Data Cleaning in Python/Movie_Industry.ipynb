```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (15,10)
```

```python
# checking the data

df = pd.read_csv('movies.csv')
df.head()
```

```python
# lets see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# datatypes for columns

df.dtypes
```

```python
df['genre'].unique()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# dropping duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.head()
```

```python
# budget high correlation
# company high correlation
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Gross Earnings')
plt.ylabel('Budet from Film')
plt.show()
```

```python
# plotting budget vs gross using seaborn

sns.regplot(x = 'budget',
            y = 'gross',
            data = df,
            scatter_kws={'color': 'red'},
            line_kws={'color': 'blue'})
plt.title('Budget vs Gross Earning')
plt.xlabel('Gross Earnings')
plt.ylabel('Budet from Film')
plt.show()
```

```python
# pearson, kendall, spearman
print(df.corr(method='pearson'))
print('-'*100)
print(df.corr(method='kendall'))
print('-'*100)
print(df.corr(method='spearman'))
```

```python
# Visualizing Correlation
i = ['pearson', 'kendall', 'spearman']
for x in i:
    correlation_matrix = df.corr(method=x)
    sns.heatmap(correlation_matrix, annot=True)
    plt.title('Correlation Matrix using {}'.format(x))
    plt.xlabel('Movie Features')
    plt.ylabel('Movie Features')
    plt.show()
```

```python
df.company.value_counts()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
# Visualizing Correlation
i = ['pearson', 'kendall', 'spearman']
for x in i:
    correlation_matrix = df_numerized.corr(method=x)
    sns.heatmap(correlation_matrix, annot=True)
    plt.title('Correlation Matrix using {}'.format(x))
    plt.xlabel('Movie Features')
    plt.ylabel('Movie Features')
    plt.show()
```

```python
df_numerized.corr()
```

```python
# unstacking to get the correlation in a readable format

correlation_mat = df_numerized.corr()
corr_pair = correlation_mat.unstack()
corr_pair
```

```python
sorted_pair = corr_pair.sort_values()
sorted_pair
```

```python
# finding the highest correlation

high_corr = sorted_pair[(sorted_pair) > 0.5]
high_corr
```
